```python
# To import the necessary libraries:
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# To ensure plots are displayed inline:
%matplotlib inline

# To specify the ticker symbols of the income stocks:
ticker = ['AAPL', 'MSFT', 'JPM']

# To download historical stock data:
df = yf.download(ticker, start='2021-01-01', end='2023-01-01')

# To save data to CSV file:
df.to_csv('income_stocks_data.csv')

# To print the first few rows of the data:
print(df.head())

# To extract the Close prices of each stock:
close_prices = df['Close']
print(close_prices.head())

# To allocate weights to the assets in our portfolio:
weights = [0.4, 0.3, 0.3]  

# To calculate daily portfolio return using closing prices:
daily_return = (close_prices * weights).pct_change()

# To calculate average return of the portfolio:
avg_portfolio_return = daily_return.mean()
print("Average Portfolio Return:", avg_portfolio_return)

# To calculate volatility of the portfolio:
portfolio_volatility = daily_return.std()
print("Portfolio Volatility:", portfolio_volatility)

# To calculate skewness of the portfolio:
portfolio_skewness = daily_return.skew()
print("Portfolio Skewness:", portfolio_skewness)

# To calculate kurtosis of the portfolio:
portfolio_kurtosis = daily_return.kurt()
print("Portfolio Kurtosis:", portfolio_kurtosis)

# To calculate portfolio Covariance Matrix:
portfolio_covariance = daily_return.cov()
print("Portfolio Covariance Matrix:")
print(portfolio_covariance)

# To calculate portfolio Correlation Matrix:
portfolio_correlation = daily_return.corr()
print("Portfolio Correlation Matrix:")
print(portfolio_correlation)
```
